```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import time
import scipy.linalg as la
from scipy.stats import zscore
import seaborn as sns
from tqdm import trange
import biosfa
import math
```

# Figure 17

```python
file = open('data.txt', 'r')
lines = file.read().splitlines()

y_original = np.array(lines).astype(np.float)

size = len(y_original)
res = []
for k in range(3,size,1):
    y = y_original[0:k]
    
    y1_lower = 0
    y1_upper = 15
    y2_lower = -1
    y2_upper = 1

    y1_abs = abs(y1_lower) + abs(y1_upper)
    y2_abs = abs(y2_lower) + abs(y2_upper)

    x = np.linspace(1, len(y), len(y))

    y1 = np.interp(y, (y.min(), y.max()), (y1_lower, y1_upper))
    y2 = np.interp(y, (y.min(), y.max()), (y2_lower, y2_upper))

    gaussian_noise = np.random.normal(0.0, y1_abs*0.1, len(x))
    y1_noise = zscore(y1 + gaussian_noise)
    gaussian_noise = np.random.normal(0.0, y2_abs*0.1, len(x))
    y2_noise = zscore(y2 + gaussian_noise)

    signal = np.zeros((k, 2))

    for i in range(len(signal)):
        signal[i][0] = y1_noise[i]
        signal[i][1] = y2_noise[i]

    signal = signal - np.mean(signal, axis=0)
    signal = signal @ la.inv(la.sqrtm(np.cov(signal.T)))

    X = np.zeros((2,k))
    X[0:2,:] = signal.T

    X_bar = X[:,0:k-1] + X[:,1:k]
    X_dot = X[:,1:k] - X[:,0:k-1]

    Cxx = np.cov(X)
    Cxx_bar = np.cov(X_bar)
    Cxx_dot = np.cov(X_dot)
    Rxx = la.sqrtm(Cxx)

    evals_bar, evecs_bar = la.eigh(Cxx_bar, Cxx)
    evals_dot, evecs_dot = la.eigh(Cxx_dot, Cxx)
    
    slow_gevp = evecs_dot[:,0] @ X
    res.append([evals_dot[0]])
```

```python
rg = np.random.default_rng(0)
iters = size

x_mean = X.mean(axis=1)
Minv, W, avg, history = biosfa.fit(
    X - x_mean[:, None],
    1,
    2e-4, 2e-4, 1e8,
    num_epochs=1,
    pre_centered=True,
    colvar=False, rg=rg, dtype=float,
)
slow_online = Minv @ W @ (X[:, -size:] - x_mean[:, None])

gevp_V = evecs_bar[:,-1:].T
gevp_V = la.inv(la.sqrtm(gevp_V @ Cxx @ gevp_V.T)) @ gevp_V
gevp_s = np.trace(gevp_V @ Cxx_dot @ gevp_V.T)

slowness_errors = []
t, Minv_t, W_t = history
for Minv_, W_ in zip(Minv_t, W_t):
    V = Minv_ * W_
    V = la.inv(la.sqrtm(V @ Cxx @ V.T)) @ V
    slowness_errors.append(np.trace(V @ Cxx_dot @ V.T))
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
plt.semilogx(res, label="Closed-form error")
plt.semilogx(t, slowness_errors, label="Biological SFA error")
plt.ylabel("Error Measure")
plt.xlabel("Sample size/iterations")
plt.grid()
plt.xlim(1,iters)
plt.legend(loc="lower left")
plt.grid()
plt.show()
```
